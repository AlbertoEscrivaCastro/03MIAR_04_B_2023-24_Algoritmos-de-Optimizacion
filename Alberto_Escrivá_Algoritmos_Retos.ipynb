## Implementa el algoritmo de la Criba de Erastótenes para encontrar número primos. Encuentra los números primos entre los 10.000 primeros números naturales

```python
# Criba de Erastótenes
import time
import numpy as np

def criba_erastotenes( n , metrica = False ):
  if metrica:
    # Inicialización para la impresión de métricas.
    tiempos = [ 0 , time.time() ]

  st_primos = set(range( 1 , n+1 ))

  for i in range( 2 , int( np.sqrt( n ) // 1 ) ):
    if i in st_primos:
      for j in range( i , int(n // i) + 1 ):
        st_primos.discard( i * j )

        if metrica:
          # Imprime las métricas por pantalla si se ha activado la opción.
          tiempos[ 0 ] = tiempos[ 1 ]
          tiempos[ 1 ] = time.time()
          print( f'Duración iteración {i},{j}: {tiempos[ 1 ] - tiempos[ 0 ]}' )

  return st_primos
```

## Implementa una solución del problema de las Torres de Hanoi para un número arbitrario n de fichas.

```python
# Torres de Hanoy
def torres_hannoi( columnas , fichas , desde , hasta ):
  # Se inicializan las variables.
  movimientos   = []

  # Por eliminación se obtiene cuál es la columna que se usará como pivote.
  pivote        = set( columnas )
  pivote.remove( desde )
  pivote.remove( hasta )
  pivote        = pivote.pop()

  # Se calculan los movimientos.
  if fichas == 1:
    movimientos = [ ( desde , hasta ) ]

  else:
    movimientos.extend( torres_hannoi( columnas , fichas-1 , desde  , pivote ) )
    movimientos.extend( torres_hannoi( columnas ,        1 , desde  , hasta  ) )
    movimientos.extend( torres_hannoi( columnas , fichas-1 , pivote , hasta  ) )

  return movimientos



# Visualiza el estado de las torres.
def visualiza_torres( torres ):
  for torre in torres:
    print( f'{torre} : {torres[ torre ]}' )



# Visualización de los movimientos.
def visualizacion_hannoi( columnas , fichas , desde , hasta ):
  tiempos = [ time.time() ]
  torres = dict()

  for columna in columnas:
    torres[ columna ] = []

  torre_ini       = list( range( 1, fichas + 1 ) )
  torre_ini.reverse()
  torres[ desde ] = torre_ini

  tiempos.append( time.time() )

  movimientos = torres_hannoi( columnas , fichas , desde , hasta )

  tiempos.append( time.time() )

  print( f'Solución calculada en: {tiempos[ 2 ] - tiempos[ 1 ]} s.' )

  visualiza_torres( torres )

  for movimiento in movimientos:
    ficha = torres[ movimiento[ 0 ] ].pop()
    torres[ movimiento[ 1 ] ].append( ficha )

    print()
    print(f'Lleva la ficha {ficha} desde {movimiento[ 0 ]} hasta {movimiento[ 1 ]}.')
    visualiza_torres( torres )

  tiempos.append( time.time() )

  print(f'Tiempo total: {tiempos[ 3 ] - tiempos[ 0 ]} s.')
```

```python
def grad_desc( funcion , Xn , gamma , tol , maxit ):
  # En caso de que el punto inicial no venga definido en un array se convierte en uno.
  try   : _       = iter( Xn )
  except: Xn      = [ Xn ]
  Xn              = sym.Array( Xn )

  # Primero se obtienen todas las variables de todas las componentes de la función y se ordenan.
  X               = list( funcion.free_symbols )
  X.sort( key = lambda xi: xi.name)

  # Se calcula el gradietne de forma simbólica, una única vez para que valga para todas
  # las iteraciones.
  gradiente       = grad_f( funcion , X )

  # Se inicializa el paso de forma que la primera vez siemproe sea mayor que la tolerancia buscada.
  paso            = 2 * tol
  iteraciones     = 0

  while paso >= tol and iteraciones <= maxit:
    iteraciones += 1

    gradiente_Xn , Xn_dict = evalua_list_func( gradiente , Xn , X )
    variacion     = gamma * gradiente_Xn
    Xn            = Xn - variacion

    try:
      paso        = np.linalg.norm( variacion )
    except:
      paso        = abs( variacion )

  if paso > tol:
    print(f"Se ha alcanzado el límite máximo de iteraciones permitido.\nEl valor obtenido es {Xn_dict}.")
  else:
    print(f"Tras {iteraciones} iteraciones se ha obtenido el resultado {Xn_dict}.")

  return Xn_dict
```

```python
def Add_Nodo(problem, H ,T ) :
  # Mejora:Establecer una funcion de probabilidad para
  # añadir un nuevo nodo dependiendo de los nodos mas cercanos y de las feromonas depositadas
  Nodos = list( problem.get_nodes() )

  # return random.choice(   list(set(range(1,len(Nodos))) - set(H) )  )
  # La línea anterior es la que se dio originalmente en el AG3, que devolvía elecciones aleatorias.
  # Es la que se reemplazará por el código necesario para basar las elecciones en la concentración de feromonas.
  J     = list( set( Nodos ) - set( H ) )
  alfa  = 5
  beta  = 5

  # Se calculan los numeradores de la expresión de las probabilidades y se colocan en una lista de ceros según su posición correspondiente,
  # sólo calculando para los nodos que faltan por visitar.
  numeradores = [0] * len(Nodos)
  for j in J:
    numeradores[j]  = ((T[H[-1]][j])**alfa)*(1/(problem.get_weight(H[-1],j))**beta)

  # Se calcula el denominador a partir de la lista completa. Al tener a 0 las posiciones de los nodos ya visitados, bastará con la suma.
  denominador = sum(numeradores)

  probabilidades = [0] * len(Nodos)
  for j in J:
    probabilidades[j] = numeradores[j] / denominador

  # Finalmente se elige el nuevo nodo. En esta implementación se emplea la lista completa de nodos pues la lista de probabilidades se encarga
  # excluir los nodos por los que ya se a pasado por al forma de construirla.
  [nodo] = random.choices( Nodos , probabilidades )

  return nodo


def Incrementa_Feromona(problem, T, H ) :
  # Incrementa segun la calidad de la solución. Añadir una cantidad inversamente proporcional a la distancia total
  for i in range(len(H)-1):
    T[H[i]][H[i+1]] += 1000/distancia_total(H, problem)
  return T

def Evaporar_Feromonas(T ):
  # Evapora 0.3 el valor de la feromona, sin que baje de 1
  # Mejora:Podemos elegir diferentes funciones de evaporación dependiendo de la cantidad actual y de la suma total de feromonas depositadas,...
  T = [[ max(T[i][j] - 0.3 , 1) for i in range(len(Nodos)) ] for j in range(len(Nodos))]
  return T
```
